###  <--- INTRODUCTION TO SEABORN--->
### Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
# Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set seaborn theme
sns.set_theme(style="whitegrid")
```

### 1> Distribution Plots : Distribution plots help you understand the distribution of a dataset.

```python
 # Example dataset
data = np.random.randn(1000)

# Histogram: Plots the frequency of data values as bars.
sns.histplot(data, kde=True, bins=30, color='skyblue')
plt.title("Histogram with KDE")
plt.show()

# KDE Plot (kdeplot): Plots a smooth curve estimating the probability density of the data.

sns.kdeplot(data, fill=True, color='red') 
plt.title("Kernel Density Estimation")
plt.show()

# distplot (deprecated): Older function for histogram with KDE

sns.histplot(data, kde=True, bins=30, color='blue')
plt.title("Histogram with KDE")
plt.show()
```

### Categorical Plots : Useful for visualizing categorical variables.

```python
# Example categorical data
categories = np.random.choice(['A', 'B', 'C'], size=200)

# Countplot: Shows the count of occurrences for each category as bars.
sns.countplot(x=categories, color='skyblue')
plt.title("Countplot Example")
plt.show()

# Barplot: Shows the average value of a numeric variable for each category.
sns.barplot(x='day', y='total_bill', data=tips, color='skyblue')
plt.title("Average Bill per Day")
plt.show()

# Boxplot: Displays the distribution of a numeric variable with median, quartiles, and outliers.
sns.boxplot(x='day', y='total_bill', data=tips, color='lightgreen')
plt.title("Boxplot Example")
plt.show()

# Violinplot: Combines a boxplot and KDE to show the distribution and density of the data.
sns.violinplot(x='day', y='total_bill', data=tips, color='pink')
plt.title("Violinplot Example")
plt.show()
```

### 4> Relational Plots : Visualize relationships between two numerical variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

tips = sns.load_dataset('tips')

# Scatterplot: Shows the relationship between two numeric variables, with optional grouping by hue.
#  hue is used to add a third variable by color
sns.scatterplot(x='total_bill', y='tip', hue='day', data=tips, palette='Set2')
plt.title("Scatterplot: Tip vs Total Bill")
plt.show()

# Lineplot: Shows trends or patterns in numeric data, often across an ordered variable.
sns.lineplot(x='size', y='total_bill', data=tips.dropna(), color='green')
plt.title("Lineplot Example")
plt.show()
```

# 5> Heatmap (numeric only) :

```python
# Heatmap : Visualizes the correlation (or values) between numeric variables using color intensity.

numeric_tips = tips.select_dtypes(include=np.number)
corr = numeric_tips.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

### Pairplot & Jointplot: 


```python
# pairplot: Shows pairwise relationships between numeric variables, optionally colored by a categorical variable.
 
sns.pairplot(tips, hue='sex', palette='husl')
plt.show()

# jointplot: Shows the relationship between two variables along with their individual distributions.

sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter', color='purple')
plt.show()
```

### TITANIC DATASET : 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
titanic = sns.load_dataset('titanic')

# Survival vs Age
sns.histplot(data=titanic, x='age', hue='survived', multiple='stack', palette='Set1')
plt.title("Survival by Age")
plt.show()

# Survival vs Class
sns.countplot(x='pclass', hue='survived', data=titanic, palette='Set2')
plt.title("Survival by Class")
plt.show()

# Survival vs Gender
sns.countplot(x='sex', hue='survived', data=titanic, palette='Set3')
plt.title("Survival by Gender")
plt.show()

# Correlation Heatmap (numeric only)
numeric_titanic = titanic.select_dtypes(include='number')
sns.heatmap(numeric_titanic.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Titanic Feature Correlation")
plt.show()
```
